```python
import json
import urllib.request
```

```python
url = (
    'https://github.com/galaxyproject/galaxy_codex/raw/e998d942f04c041b0ae8282f0380036431fbbde2'
    '/communities/imaging/resources/tools_filtered_by_ts_categories.json'
)
filename = 'tools.json'
urllib.request.urlretrieve(url, filename)
```

```python
with open(filename) as fin:
    tools = json.load(fin)
```

```python
def sum_runs_on_main_servers(tool):
    s = 0
    for key, value in tool.items():
        if key.startswith('Suite runs (usegalaxy.'):
            s += value
    return s
```

```python
workflows = set()
counters = dict()
min_unique_users_eu = 0
min_unique_users_main = 0

for tool in tools:
    if any(
        [
            'bmcv' in (tool[key].lower() if tool[key] else '')
            for key in ['Homepage', 'Suite source', 'Suite parsed folder']
        ]
    ):
        print(tool['Suite ID'])
        for key in ['Suite runs (usegalaxy.eu)', 'Suite runs on main servers']:
            counters.setdefault(key, 0)
            counters[key] += tool[key]
        assert sum_runs_on_main_servers(tool) == tool['Suite runs on main servers']
        workflows |= set(tool['Related Workflows'])
        min_unique_users_eu = max((min_unique_users_eu, tool['Suite users (usegalaxy.eu)']))
        min_unique_users_main = max((min_unique_users_main, tool['Suite users on main servers']))

workflows_wfh = list(filter(lambda url: 'workflowhub' in url.lower(), workflows))
print()
print('Number of workflows using GIA tools:', len(workflows))
print('Number of those on WorkflowHub:', len(workflows_wfh))
print('\n'.join([f'- {url}' for url in workflows_wfh]))

print()
for key, value in counters.items():
    print(f'{key}: {value}')
    
print()
print(f'Unique users on EU: ≥{min_unique_users_eu}')
print(f'Unique users on main: ≥{min_unique_users_main}')
```
